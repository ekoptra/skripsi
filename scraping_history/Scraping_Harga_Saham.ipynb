```python
import requests
import datetime
import pandas as pd
import os
```

Periode 1 merupakan tanggal awal dan periode 2 merupakan tanggal akhir. Tanggal harus berbentuk timestamp sehingga perlu dikonversi terlebih dahulu dan data di yahoo finance juga tersedia dari jam 8 pagi

```python
periode1 = "01-01-2000 08"
periode1 = datetime.datetime.strptime(periode1, "%d-%m-%Y %H")
periode1 = int(datetime.datetime.timestamp(periode1))
periode1
```

```python
periode2 = datetime.datetime.now()
periode2 = periode2.strftime("%d-%m-%Y") + " 08"
periode2 = datetime.datetime.strptime(periode2, "%d-%m-%Y %H")
periode2 = int(datetime.datetime.timestamp(periode2))
periode2
```

```python
list_saham = ["ADRO", "ASII", "BBNI", "BBRI",
             "BMRI", "EXCL", "GGRM", "ICBP", 
             "KLBF", "LPPF", "PGAS", "SMGR",
             "SSMS", "TLKM", "UNVR"]

if not os.path.exists('data/harga_saham'):
        os.makedirs('data/harga_saham')
        
for saham in list_saham:
    link = f"https://query1.finance.yahoo.com/v7/finance/download/{saham}.JK?period1={periode1}&period2={periode2}&interval=1d&events=history&includeAdjustedClose=true"
    page = requests.get(link, headers={'user-agent': 'My app'})
    open(f'data/harga_saham/{saham}.csv', 'wb').write(page.content)
```

```python
tanggal_awal = []
for saham in list_saham:
    data = pd.read_csv(f"data/harga_saham/{saham}.csv")
    tanggal_awal.append({"saham" : saham, "awal" : data.Date[0], "jumlah_data" : data.shape[0]})
    
pd.DataFrame(tanggal_awal).sort_values("awal")
```
